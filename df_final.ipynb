```python
# Importing the required libraries
import pandas as pd
import numpy as np
```

```python
#uploading the data set
dfIH=pd.read_csv('D:/Invoice Headers.csv')
dfID=pd.read_csv('D:/Invoice Detail.csv')
dfCooP=pd.read_csv('D:/CooP.csv')
```

```python
#from the business problem I understand that there are 

#1. Unit Cost
#2. Rebate In Agreement Currency 
#3. PO
#4. Asin 
#5. Quantity Recieved     
```

```python
#First step is to join the tables 

# In CooP Column the correct the column names from Purchase Order to PO and Asin to ASIN 

# Renaming columns in CooP
dfCooP = dfCooP.rename(columns={'Purchase Order': 'PO', 'Asin': 'ASIN'})


```

```python
# View specific columns
selected_columns = dfCooP[['PO', 'ASIN']]
print(selected_columns)
```

```python
# merge the tables CooP & Invoice Detail with PO and ASIN
df = pd.merge(dfID, dfCooP, on=['PO', 'ASIN'])
```

```python
df.head(20)
```

```python
#now data cleansing for the merged table named df
```

```python
# List of columns to keep
columns_to_keep = ['Invoice Number','PO','Unit Cost', 'ASIN', 'Quantity received','Rebate In Agreement Currency']

# Delete columns not in the list

df1 = df[columns_to_keep]
```

```python
df1.head()
```

```python
# Check for missing values 
missing_values = df1.isnull().sum()

# Print the count of missing values for each column
print("Missing Values:")
print(missing_values)
```

```python
# Drop rows with 'Rebate In Agreement Currency' missing values
dfcleaned = df1.dropna(subset=['Rebate In Agreement Currency'])
print(dfcleaned)
```

```python
# joining dfcleaned with Invoice Headers 
df_final = pd.merge(dfIH, dfcleaned, on=['Invoice Number'])
```

```python
print(df_final)
```

```python
df_final.head()
```

```python

output_path = 'D:/df_final.csv'
df_final.to_csv(output_path, index=False)


```
